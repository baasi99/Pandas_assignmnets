## Q1-Concatenate the two weeks of sales data into one DataFrame. Assign the week  DataFrame a key of "Week 1" and the week2 DataFrame a key of "Week 2".

```python
import pandas as pd
```

```python
week1=pd.read_csv('c:/pandas_docs/week_1_sales.csv')
```

```python
week2=pd.read_csv('c:/pandas_docs/week_2_sales.csv')
```

```python
week1
```

## Q2 Find the customers who ate at the restaurant both weeks.

```python
wk=week1.merge(week2,how='inner',on='Food ID')
```

```python
wk
```

```python
cust=pd.read_csv('c:/pandas_docs/customers.csv')
```

```python
a=wk.merge(cust,how='left', left_on='Food ID', right_on='ID',indicator=True)
```
